```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer
```

Create Twitter API module from tweepy by using client ID & secret.

```python
consumer_key = 'YOUR CONSUMER KEY'
consumer_secret = 'YOUR CONSUMER SECRET'
access_token = 'YOUR ACCESS TOKEN'
access_secret = 'YOUR ACCESS SECRET'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

tweet_ids = tw.tweet_id.values
len(tweet_ids)
```

Get detail of each tweet by its tweed id from tweepy and store in txt json file.

```python
count = 0
fails_dict = {}
start = timer() 

with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))        
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start) 
print(fails_dict)
```

Import the created json file into a pandas dataframe.

```python
rt_list = []
with open("tweet_json.txt") as f:
    for line in f:
        rt_list.append(json.loads(line))

#rt_list
rt_extended = pd.DataFrame.from_records(rt_list)
rt = rt_extended[['id', 'retweet_count', 'favorite_count']]
rt.info()
```

## Clean

```python
tw_clean = tw.copy()
im_clean = im.copy()
rt_clean = rt.copy()
```

### Tidiness

#### One variable in four columns in `twitter-archive-enhanced` table (*doggo*, *floofer*, *pupper*, *puppo*)

##### Define
Create a new column *stage* and set its value to one of these: *"doggo"*, *"floofer"*, *"pupper"*, *"puppo"* or *"None"* if not specified. Then categorize the *stage* column.

##### Code

```python
conditions = [
    (tw_clean['doggo'] == 'doggo'),
    (tw_clean['floofer'] == 'floofer'),
    (tw_clean['pupper'] == 'pupper'),
    (tw_clean['puppo'] == 'puppo')]
choices = ['doggo', 'floofer', 'pupper', 'puppo']
tw_clean['stage'] = np.select(conditions, choices, default='None')
tw_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
tw_clean.state = tw_clean.stage.astype('category')
```

##### Test

```python
for index, row in tw_clean.iterrows():
    if row['stage'] != 'None':
        assert(row['stage'] in row['text'].lower())
```

#### `rt` table, which contains retweet_count and favorite_count, should be part of the `twitter-archive-enhanced` table .

##### Define
Merge the *retweet_count* and *favorite_count* columns to the `twitter-archive-enhanced` table, joining on *tweet_id*.

##### Code

```python
rt_clean.rename(columns={'id': 'tweet_id'}, inplace=True)
tw_clean = pd.merge(tw_clean, rt_clean,
                            on=['tweet_id'], how='left')
tw_clean = tw_clean.dropna(subset=['retweet_count', 'favorite_count'])
tw_clean.retweet_count = tw_clean.retweet_count.astype(int)
tw_clean.favorite_count = tw_clean.favorite_count.astype(int)
```

##### Test

```python
tw_clean.head()
```

### Quality

#### *timestamp* is string not datetime

Convert the *timestamp* column's data type from string to datetime using `astype`.

#### *name* is entered as *"a"* sometimes instead of *None*

#### 59 entries do not have images

Apply a function that finds the tweet rows without images and drop them from the `twitter-archive-enhanced` table.

#### inaccurate rating_denominator values (different than 10)

#### Define
Find tweets having *rating_denominator* values different than 10 and drop them from the `twitter-archive-enhanced` table.

```python
tw_clean = tw_clean.drop(tw_clean[tw_clean.rating_denominator != 10].index)
```

##### Test

```python
tw_clean.rating_denominator.value_counts()
```

#### All lowercase p1, p2, p3 sometimes, first letter is uppercase other times.

#### Define
Convert *p1*, *p2*, *p3* column values to all lowercase letters using *lower* function.

#### Code

```python
im_clean["p1"] = im_clean["p1"].str.lower()
im_clean["p2"] = im_clean["p2"].str.lower()
im_clean["p3"] = im_clean["p3"].str.lower()
```

#### Test

```python
for index, row in im_clean.iterrows():
    assert(any(letter.isupper() for letter in row["p1"]) == False)
    assert(any(letter.isupper() for letter in row["p2"]) == False)
    assert(any(letter.isupper() for letter in row["p3"]) == False)
```

#### *p1*, *p2*, *p3* containing breaks use underscore ("_") sometimes, dash ("-") other times.

#### Define
Convert dashes ("-") existing in *p1*, *p2*, *p3* columns to underscores ("_") using *replace* function.

#### Code

```python
im_clean["p1"] = im_clean["p1"].str.replace("-", "_")
im_clean["p2"] = im_clean["p2"].str.replace("-", "_")
im_clean["p3"] = im_clean["p3"].str.replace("-", "_")
```

#### Test

```python
for index, row in im_clean.iterrows():
    assert("-" not in row["p1"])
    assert("-" not in row["p2"])
    assert("-" not in row["p3"])
```

#### duplicated *expanded_url* values

#### Define
Find tweet rows dith duplicated *expanded_url* columns by usinf *duplicated* function and drop them from the `twitter-archive-enhanced` table.

#### Code

```python
tw_clean.drop_duplicates(subset='expanded_urls', inplace=True)
```

### Storing cleaned dataframes

In the beginning we had three seperate datasets. Due to tidiness issues, we have merged the `rt` dataframe to `tw` dataframe as each type of observational unit forms a table.

I did not merge `im` dataframe to `tw` dataframe because each dataset holds different types of observational units for each tweet of WeRateDogs Twitter account. In `tw` dataframe, we hold every statistical and informative data of a tweet such as the favourite count, tweet text, name and breed of the dog mentioned in the tweet. On the other hand the `im` dataframe contains the dog breed predictions of each dog mentioned in each WeRateDogs Twitter account by using neutral network, which is totally a different type of observation.
